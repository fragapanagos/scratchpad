```python
from numpy import *
from matplotlib.pyplot import *
```

```python
def get_rms(s):
    return sqrt(sum(s**2)/s.shape[0])
```

```python
def genBandlimWGN_gen(T, rng, dt=0.001, rms=.5, bandwidth=[0, 10], dim=1):
    assert dt>0
    assert type(dim) is int and dim>0
    assert (bandwidth[0] >=0) and (bandwidth[0] <= bandwidth[1]) and (bandwidth[1] < 1./(2*dt))

    N = int(T/dt)
    Fs = 1./dt

    WGN = randn(N,dim)
    figure('WGN generator', figsize=(8,6)).clf()
    subplot(221)
    t = arange(0,T,dt)
    plot(t, WGN)
    title('WGN')
    
    freq = 2 * pi * arange(N, dtype=float) / N
    freq = unwrap(fftshift(freq) - 2*pi)
    freq = freq * Fs/(2*pi)

    S = fftshift(fft.fft(WGN, axis=0))
    
    subplot(222)
    plot(freq, abs(S))
    xlim([-Fs/2, Fs/2])
    title('Spectrum of WGN')
    
    #find the part of the axis we're interested in
    zero_index = logical_or(abs(freq)<bandwidth[0], abs(freq)>bandwidth[1])
    S[zero_index,:] = 0
    subplot(224)
    plot(freq, abs(S))
    xlim([-Fs/2, Fs/2])
    xlabel('Frequency (Hz)')
    title('Spectrum of WGN after filtering')
    
    WGN_BL = ifft(ifftshift(S, axes=0), axis=0).real #get time domain signal
    rmsS = get_rms(WGN_BL) # scale rms
    WGN_BL = WGN_BL*(rms/rmsS)
    subplot(223)
    plot(t, WGN_BL)
    xlabel('Time (s)')
    title('Bandlimited WGN')
    
    return WGN_BL
```

```python
def plotSpectrum(s, dt):
    Fs = 1./dt
    N = len(s)
    freq = 2 * pi * arange(N, dtype=float) / N
    freq = unwrap(fftshift(freq) - 2*pi)
    freq = freq * Fs/(2*pi)
    S = fftshift(fft.fft(s, axis=0))
    
    figure('spectrum')
    plot(freq, abs(S))
    xlim([-Fs/2, Fs/2])
    xlabel('Frequency (Hz)')
```

```python
def genBandlimWGN_inv(T, rng, dt=0.001, rms=.5, bandwidth=[0, 10], dim=1):
    assert dt>0
    assert type(dim) is int and dim>0
    assert (bandwidth[0] >=0) and (bandwidth[0] <= bandwidth[1]) and (bandwidth[1] < 1./(2*dt))

    N = int(T/dt)
    Fs = 1./dt

    figure('WGN generator', figsize=(8,6)).clf()
   
    freq = arange(N, dtype=float) * Fs / N   

    #find the part of the axis we're interested in
    nonzero_index = logical_and(abs(freq)>=bandwidth[0], abs(freq)<=bandwidth[1])
    numnonzero = len(where(nonzero_index)[0])

    S = zeros((N,dim), dtype=complex)
    S[nonzero_index,:] = rng.randn(numnonzero, dim) + 1j*rng.randn(numnonzero, dim)
    S[0,:] = abs(S[0,:])
    
    if N%2==0:
        S[N/2+1:,:] = flipud(S[1:N/2,:].conj())
    else:
        S[N/2+1:,:] = flipud(S[1:N/2+1,:].conj())
    
    subplot(121)
    plot(freq, abs(S),'o-')
    xlim([0, Fs])
    axvline(Fs/2., ls=':')
    xlabel('Frequency (Hz)')
    title('Spectrum of WGN')
    
    t = arange(N)*T/N
    WGN_BL = ifft(S, axis=0).real #get time domain signal
    rmsS = get_rms(WGN_BL) # scale rms
    WGN_BL = WGN_BL*(rms/rmsS)
    subplot(122)
    plot(t, WGN_BL)
    xlabel('Time (s)')
    title('Bandlimited WGN')
    
    return WGN_BL
```

```python
#Test the invfilter BL WGN generator
rng = random.RandomState()
T = .5
dt = .0013
rms = .5
bandwidth = [0, 10]
sig = genBandlimWGN_inv(T, rng, rms=rms, dt=dt, bandwidth=bandwidth)

plotSpectrum(sig, dt)
```
