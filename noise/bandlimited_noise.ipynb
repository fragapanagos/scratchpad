```python
from numpy import *
from matplotlib.pyplot import *
```

```python
def get_rms(s):
    return sqrt(sum(s**2)/s.shape[0])
```

```python
def genBandlimWGN(T, rng, dt=0.001, rms=.5, bandwidth=[0, 10], dim=1):
    assert dt>0
    assert type(dim) is int and dim>0
    assert (bandwidth[0] >=0) and (bandwidth[0] <= bandwidth[1]) and (bandwidth[1] < 1./(2*dt))

    N = floor(T/dt)
    Fs = 1./dt

    WGN = randn(N,dim)
    figure('WGN generator', figsize=(8,6)).clf()
    subplot(221)
    t = arange(0,T,dt)
    plot(t, WGN)
    title('WGN')
    
    freq = 2 * pi * arange(N, dtype=float) / N
    freq = unwrap(fftshift(freq) - 2*pi)
    freq = freq * Fs/(2*pi)

    S = fftshift(fft.fft(WGN, axis=0))
    
    subplot(222)
    plot(freq, abs(S))
    xlim([-Fs/2, Fs/2])
    title('Spectrum of WGN')
    
    #find the part of the axis we're interested in
    zero_index = logical_or(abs(freq)<bandwidth[0], abs(freq)>bandwidth[1])
    S[zero_index,:] = 0
    subplot(224)
    plot(freq, abs(S))
    xlim([-Fs/2, Fs/2])
    xlabel('Frequency (Hz)')
    title('Spectrum of WGN after filtering')
    
    WGN_BL = ifft(ifftshift(S, axes=0), axis=0).real #get time domain signal
    rmsS = get_rms(WGN_BL) # scale rms
    WGN_BL = WGN_BL*(rms/rmsS)
    subplot(223)
    plot(t, WGN_BL)
    xlabel('Time (s)')
    title('Bandlimited WGN')
    
    return WGN_BL
```

```python
def plotSpectrum(s, dt):
    Fs = 1./dt
    N = len(s)
    freq = 2 * pi * arange(N, dtype=float) / N
    freq = unwrap(fftshift(freq) - 2*pi)
    freq = freq * Fs/(2*pi)
    S = fftshift(fft.fft(s, axis=0))
    
    figure('spectrum')
    plot(freq, abs(S))
    xlim([-Fs/2, Fs/2])
    xlabel('Frequency (Hz)')
```

```python
#Test the BL WGN generator
rng = random.RandomState()
T = 10.
dt = .001
rms = .5
bandwidth = [0, 10]
sig = genBandlimWGN(T, rng, rms=rms, dt=dt, bandwidth=bandwidth)
print get_rms(sig)

plotSpectrum(sig, dt)
```
