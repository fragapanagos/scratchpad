This notebook looks into simulating a continuous time low pass filter digitally.

```python
import numpy as np
import scipy.signal as sig
```

```python
def plotSpectrum(s, dt):
    Fs = 1./dt
    N = len(s)
    freq = 2 * pi * arange(N, dtype=float) / N
    freq = unwrap(fftshift(freq) - 2*pi)
    freq = freq * Fs/(2*pi)
    S = fftshift(fft.fft(s, axis=0))
    
    figure('spectrum')
    plot(freq, abs(S))
    xlim([-Fs/2, Fs/2])
    xlabel('Frequency (Hz)')
```

```python
def bode(b, a, fs):
    nyq_rate = fs/2.
    figure()
    w, h = sig.freqz(b, a, worN=4096)
    subplot(211)
    semilogx((w/pi)*nyq_rate, 20*log10(abs(h)), linewidth=2)
    ylim([-40,1])
    title('Frequency Response')
    xlabel('Frequency (Hz)')
    grid(True)
    ylabel('Gain')
    axhline(-3)
    axvline(10)
    #ylim((-20,1))
    subplot(212)
    semilogx((w/pi)*nyq_rate, np.angle(h)/pi * 180., linewidth=2)
    #semilogx((w/pi)*nyq_rate, 10*log10(np.absolute(h)), linewidth=2)
    #plot((w/pi)*nyq_rate, np.absolute(h), linewidth=2)
    grid(True)
```

```python
T = 1.
dt = .001
fs = 1./dt
tau = .1

t = np.arange(0,T,dt)
#x = np.cos(2*np.pi*t)
#x = np.random.randn(len(t))
x = np.ones(len(t))
x[0] = 1
```

```python
# using scipy bilinear
b,a = sig.bilinear([1.], [1.,tau], fs)
#b,a = sig.bilinear([1.], [1.,tau], fs*np.pi/180.)
print b
print a

bode(b, a, fs)
subplot(211)
#ylim([-20,1])

xfilt = sig.lfilter(b, a, x)
figure('scipy filtering')
subplot(211)
plot(t,x)
subplot(212)
plot(t,xfilt); tight_layout()
```

```python
# using hand calculated bilinear transformation
#b = np.array([1,1]) / (1+2*tau/dt)
#a = np.array([1+2*tau/dt, 1-2*tau/dt]) / (1+2*tau/dt)
b = np.array([1,1])
a = np.array([1+2*tau/dt, 1-2*tau/dt])
print b
print a

bode(b, a, fs)
xfilt = sig.lfilter(b, a, x)
figure('manual filtering')
subplot(211)
plot(t,x)
subplot(212)
plot(t,xfilt); tight_layout()
```

```python
# using matlab filter design toolbox
b = [.03, .03]
a = [1., -.939]
print b
print a

bode(b, a, fs)
xfilt = sig.lfilter(b, a, x)
figure('matlab filtering')
subplot(211)
plot(t,x)
subplot(212)
plot(t,xfilt); tight_layout()
```

```python
# using what I know about discretized differential equations 
b = [1-np.exp(-dt/tau)]
a = [1, -np.exp(-dt/tau)]
print b
print a

xfilt = sig.lfilter(b, a, x)

bode(b, a, fs)


figure('filtering')
subplot(211)
plot(t,x)
subplot(212)
plot(t,(1-np.exp(-t/tau)),'x')
plot(t,xfilt, 'r')
axvline(tau)
axhline(1-np.exp(-1))

plotSpectrum(x, dt)
xlim(-20, 20)
```
